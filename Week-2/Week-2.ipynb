```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_oscillators = 100  
t_max = 100  
dt = 0.1
time = np.arange(0, t_max, dt)

frequencies = np.random.normal(0, 1, n_oscillators)

def kuramoto(coupling_strength):
    phases = np.random.uniform(0, 2 * np.pi, n_oscillators)
    phase_history = []
    
    for t in time:
        phase_history.append(phases.copy())
        for i in range(n_oscillators):
            coupling = 0
            for j in range(n_oscillators):
                coupling += np.sin(phases[j] - phases[i])
            
            phases[i] += (frequencies[i] + (coupling_strength / n_oscillators) * coupling) * dt
    
    return np.array(phase_history)

def calculate_r(phases):
    complex_mean = np.mean(np.exp(1j * phases), axis=1)
    return np.abs(complex_mean)

plt.figure(figsize=(12, 4))

# Weak Coupling
phases_weak = kuramoto(0.1)
sync_weak = calculate_r(phases_weak)
plt.subplot(121)
plt.plot(time, sync_weak, 'b-', label='r')
plt.title('Weak Coupling (K=0.1)')
plt.xlabel('Time')
plt.ylabel('r')
plt.ylim(0, 1)
plt.grid(True)

# Strong Coupling
phases_strong = kuramoto(10.0)
sync_strong = calculate_r(phases_strong)
plt.subplot(122)
plt.plot(time, sync_strong, 'r-', label='r')
plt.title('Strong Coupling (K=10.0)')
plt.xlabel('Time')
plt.ylabel('r')
plt.ylim(0, 1)
plt.grid(True)

plt.tight_layout()
plt.savefig("ku_coupling.png")
plt.show()

# Plot synchronization vs. coupling strength
coupling_range = np.linspace(0, 10, 50)
final_sync = []

for k in coupling_range:
    phases = kuramoto(k)
    sync = calculate_r(phases)
    final_sync.append(sync[-1])

plt.figure(figsize=(8, 4))
plt.plot(coupling_range, final_sync, 'k.-')
plt.xlabel('Coupling Strength (K)')
plt.ylabel('Final Synchronization')
plt.title('Synchronization vs Coupling Strength')
plt.grid(True)
plt.savefig("ku_rvsk.png")
plt.show()
```
